```python
from typing import Dict, NamedTuple


class RunKey(NamedTuple):
    """
    Represents a parameter tuple.
    """

    cost: float
    cost_threshold: float
    trust_threshold: float
    num_groups: int
    num_tags: int


def generate_runs_dict(fname: str) -> Dict[RunKey, str]:
    """
    Generate a dictionary of runs by parsing the specified file.

    Parameters
    ----------
    fname: str
        The filename which contains the run information.

    Returns
    -------
    dict
        A dictionary mapping parameter values to a distinct run id.
    """
    runs = {}
    with open(fname) as file:
        for line in file:
            if line.startswith(";"):
                continue

            items = line.split()
            run_id = items[0][:-1]
            cost, cost_thres, trust_thres = [float(items[x]) for x in [5, 6, 7]]
            num_groups, num_tags = [int(items[x]) for x in [2, 9]]

            run_key = RunKey(cost, cost_thres, trust_thres, num_groups, num_tags)
            runs[run_key] = run_id

    return runs


runs = generate_runs_dict("../results/runs.sorted")
```

```python
# timeseries plots for the evolution of cooperation
# for clarity, we present only the results every 50th timestep
palette = sns.color_palette("Set1", 10)


def timestep_plotter(df, ax, main_ticks, twin_ticks, xlabel, legend):
    plot_args = {"data": df, "x": "timestep", "legend": False}
    sns.lineplot(
        y="cooperators", label="cooperation", ax=ax, color=palette[0], **plot_args
    )
    ax.set_ylim(0, 100)

    ax.set_xlabel("timesteps" if xlabel else "")
    ax.set_ylabel("\% of population" if main_ticks else "")
    if not main_ticks:
        ax.tick_params(left=False, length=0)
        ax.set(yticklabels=[])

    line2 = ax.twinx()
    sns.lineplot(
        y="trust", label="global trust", ax=line2, color=palette[1], **plot_args
    )
    line2.set_ylim(0, 1)
    line2.grid(False)

    line2.set_ylabel("global trust" if twin_ticks else "")
    if not twin_ticks:
        line2.tick_params(right=False)
        line2.set(yticklabels=[])

    # legend
    if legend:
        ax.figure.legend(
            ncol=2, bbox_to_anchor=(0.5, 1.0), loc="lower center", frameon=False
        )


def get_filtered_df(df, run):
    _df = df.loc[run, "cooperators":"trust"].reset_index()[::50]
    _df["cooperators"] = _df["cooperators"] / 10
    _df["trust"] = _df["trust"] / 100
    return _df


rs = [0.1, 0.5]
num_runs = 2
fig, axes = plt.subplots(nrows=len(rs), ncols=num_runs, figsize=(12, 6))

for col in range(len(rs)):
    key = RunKey(rs[col] * 10, 0.5, 0.5, 100, 2)
    run_id = runs[key]
    df = pd.read_parquet(f"../results/data/model/{run_id}_model.parquet")

    for row in range(num_runs):
        run = random.choice(range(51))
        _df = get_filtered_df(df, run)
        timestep_plotter(
            df=_df,
            ax=axes[row][col],
            main_ticks=col == 0,
            twin_ticks=col == 1,
            xlabel=row == len(rs) - 1,
            legend=(row == 0 and col == 0),
        )
        if row == 0:
            axes[row][col].set_title(f"r = {rs[col]}")
        axes[row][col].grid(True)

plt.savefig(plot_dir / "cooperation-trust-timeseries.pdf", bbox_inches="tight")
plt.show()
# df = pd.read_parquet(f'../results/data/agents/{run_id}_agents.parquet')
# sns.boxenplot(x=df.loc[RUN, 'total payoff'], ax=box)
```

```python
# impact of group sizes on level of cooperation & trust
def plot_cooperation_trust(df, title, main_ticks, twin_ticks, ax):
    plot_args = {"data": df, "x": "r", "palette": "dark", "legend": False}
    sns.lineplot(y="cooperators", label="cooperation", ax=ax, color="b", **plot_args)
    ax.set_ylim(0, 100)
    ax.set_title(title)
    ax.grid(True)

    ax.set_ylabel("\% ofpopulation" if main_ticks else "")
    if not main_ticks:
        ax.tick_params(left=False, length=0)
        ax.set(yticklabels=[])

    ax1 = ax.twinx()
    sns.lineplot(y="trust", label="trust", ax=ax1, color="r", **plot_args)
    ax1.set_ylim(0, 1)

    ax1.set_ylabel("systemwide trust" if twin_ticks else "")
    if not twin_ticks:
        ax1.tick_params(right=False)
        ax1.set(yticklabels=[])

    # need to hide the grid, otherwise the grid and lines cut each other
    ax1.grid(False)

    if main_ticks:
        ax.figure.legend(
            ncol=2, bbox_to_anchor=(0.5, 1.0), loc="lower center", frameon=False
        )


df = (
    model_df.loc[(slice(None), 0.5, 0.5, slice(None), 2), ["cooperators", "trust"]]
    .reset_index()
    .drop(columns=["tt", "ct", "num_tags", "run"])
)

SIZES = [10, 50, 100]
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))
for i, (gs, ax) in enumerate(zip(SIZES, axes)):
    plot_cooperation_trust(
        df[df["group_size"] == gs], f"groupsize = {gs}", i == 0, i == len(axes) - 1, ax
    )

fig.tight_layout()
for ax in fig.axes:
    ax.set_xlabel("\\textit{r}")

plt.savefig(
    plot_dir / "impact-of-initial-group-sizes-on-cooperation-trust.pdf",
    bbox_inches="tight",
)
plt.show()
```

```python
def plot_group_size_cooperators(sizes, cooperators, cmap, ax):
    norm = matplotlib.colors.Normalize(vmin=0, vmax=100)
    colors = [cmap(norm(value)) for value in cooperators]

    tmap = squarify.plot(
        sizes=sizes,
        color=colors,
        alpha=0.7,
        ax=ax,
        bar_kwargs={"linewidth": 0.75, "edgecolor": "#000000"},
    )
    tmap.set_xticks([])
    tmap.set_yticks([])


cmap = sns.color_palette("coolwarm", as_cmap=True)
# fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(10,3))
fig = plt.figure(figsize=(10, 3))
grid = AxesGrid(
    fig,
    111,
    nrows_ncols=(1, 3),
    axes_pad=0.25,
    cbar_mode="single",
    cbar_location="right",
    cbar_pad=0.25,
    cbar_size="5%",
)

for (run, ax) in zip([10, 20, 30], grid):
    _df = group_df.loc[(0.5, 0.5, 0.5, 20, 10, run), "size":"cooperators"]
    plot_group_size_cooperators(_df["size"], _df["cooperators"], cmap, ax)

img = plt.imshow([group_df["cooperators"]], cmap=cmap)
img.set_visible(False)
cbar = ax.cax.colorbar(img)

# label for the color bar
cax = grid.cbar_axes[0]
axis = cax.axis[cax.orientation]
axis.label.set_text("% of population")

plt.savefig(plot_dir / "groupsize-vs-level-of-cooperation.pdf", bbox_inches="tight")
plt.show()
```
